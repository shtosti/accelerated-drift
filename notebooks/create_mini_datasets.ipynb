```python
import json
import pandas as pd
```

# Medarxiv

```python
datapath_medarxiv = "./../data/processed/medarxiv.jsonl"

with open(datapath_medarxiv, "r") as f:
    for line in f:
        paper = json.loads(line)
        print(paper.keys())
        break
```

```python
# create a mini dataset jsonl file with only 15 papers per month and year
# year col: "year"
# month col: "publicationDate" (format: "YYYY-MM-DD")
mini_datapath_medarxiv = "./../data/processed/medarxiv_mini.jsonl"
with open(datapath_medarxiv, "r") as f_in, open(mini_datapath_medarxiv, "w") as f_out:
    papers = []
    for line in f_in:
        paper = json.loads(line)
        papers.append(paper)
    df = pd.DataFrame(papers)
    df["year"] = df["publicationDate"].str[:4]
    df["month"] = df["publicationDate"].str[5:7]
    mini_df_medarxiv = df.groupby(["year", "month"]).head(15)
    mini_df_medarxiv.to_json(f_out, orient="records", lines=True)


# load the mini dataset into a pandas dataframe
mini_df_medarxiv = pd.read_json(mini_datapath_medarxiv, lines=True)
print(mini_df_medarxiv.head())
```
