```python
import pandas as pd
import json
from pathlib import Path


features_path = Path("your_features.csv")
debug_path = Path("syntax_debug.jsonl")

# your feature table
df = pd.read_csv(features_path)

# your syntactic debug file

debug = []
with debug_path.open() as f:
    for line in f:
        debug.append(json.loads(line))

debug_df = pd.DataFrame(debug)
```

```python
merged = df.merge(
    debug_df,
    left_index=True,
    right_on="doc_id",
    how="inner"
)
```

```python
top_entropy = merged.sort_values("dependency_entropy", ascending=False).head(5)
top_entropy[["dependency_entropy", "deepest_sentence"]]
```

```python
top_depth = merged.sort_values("clause_depth", ascending=False).head(5)
top_depth[["clause_depth", "deepest_sentence"]]
```

```python
top_dep = merged.sort_values("dependency_length", ascending=False).head(5)
top_dep[["dependency_length", "long_dependency_examples"]]
```

```python
doc = top_entropy.iloc[0]

print("ENTROPY:", doc["dependency_entropy"])
print("\nDEEPEST SENTENCE:\n", doc["deepest_sentence"]["sentence"])

print("\nLONG DEPENDENCIES:")
for ex in doc["long_dependency_examples"]:
    print("-", ex["token"], "→", ex["head"], "|", ex["sentence"])
```
